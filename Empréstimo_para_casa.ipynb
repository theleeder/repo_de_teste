Criação de um programa para aprovar o empréstimo bancário para comprar uma casa.
Será coletado o valor da casa, salário do cliente e quantos anos ele vai pagar.

A prestação mesal não pode ser maior que 30% do salário.

```python

# Entradas do usuário
casa = int(input("Valor da casa: R$"))
salario = int(input("Salário do cliente: R$"))
anos = int(input("Duração do pagamento (anos): "))


print("------------------------------") # separador visual


qtd_prest = 12 * anos           # Calcula quantas parcelas mensais terá o pagamento
prestacoes = casa / qtd_prest   # Calcula o valor de cada prestação da casa
max_prest = salario * 0.3       # Calcula 30% do salário do cliente

print(f"Número de prestações: {qtd_prest} parcelas")
print(f"Valor de cada parcela: R${prestacoes:.2f}")




print("------------------------------") # separador visual



# Resultado
if prestacoes < max_prest:
    print("Aprovado!")
else:
    print("Negado")


```
